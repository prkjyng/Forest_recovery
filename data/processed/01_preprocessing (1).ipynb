```python
import os
import pandas as pd

# 가상 환경 내의 폴더 경로 설정
raw_path = "./data/raw/"

files = {
    "검봉": "gumbong_flux.csv",
    "가리왕": "gariwang_flux.csv",
    "매화": "maehwa_flux.csv"
}

dfs = {}
for name, file_name in files.items():
    # 폴더 경로와 파일 이름을 합쳐서 올바른 경로를 만듭니다 (예: ./data/raw/gumbong_flux.csv)
    full_path = os.path.join(raw_path, file_name)

    # file_path 대신 생성한 full_path를 사용합니다.
    df = pd.read_csv(full_path, encoding="utf-8-sig")
    df["지점"] = name
    dfs[name] = df
    print(f"{name}: {df.shape} / 결측치 비율: {df.isnull().mean().mean():.2%}")

# 합치기
df_all = pd.concat(dfs.values(), ignore_index=True)
print(f"\n전체: {df_all.shape}")
print(df_all.columns.tolist())
```

```python
import numpy as np
import pandas as pd

# 1. 이전 단계에서 합친 df_all을 df로 가져와서 정렬합니다.
df = df_all.sort_values(["지점", "관측일시"]).reset_index(drop=True)

num_cols = [
    "토양수분1", "토양수분2",
    "잠열 플럭스", "현열 플럭스(보정된 온도)",
    "co2 플럭스", "순복사량",
    "기온 10미터", "습도 10미터",
    "강우량(군락 위)", "지중온도1", "지중온도2",
    "지중 열류량1", "지중 열류량2",
    "퀀텀량(군락 위)"
]

# 2. 숫자형 데이터로 변환 (문자열 등이 섞여있으면 NaN 처리)
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# 3. 지점별 선형 보간법(Linear Interpolation)으로 결측치 채우기
df[num_cols] = df.groupby("지점")[num_cols].transform(
    lambda x: x.interpolate(method="linear")
)

# 4. IQR 방식을 이용한 이상치(Outlier) 제거 (안전한 마스킹 방식)
# 모든 컬럼에 대해 정상 범위에 있는지 체크하는 조건을 만듭니다.
valid_mask = pd.Series(True, index=df.index)

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # 극단적 이상치(Extreme Outliers: 3 * IQR) 조건
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR

    # 현재 컬럼에서 정상 범위에 있는 행들만 골라 마스크 업데이트
    valid_mask = valid_mask & (df[col] >= lower_bound) & (df[col] <= upper_bound)

# 최종적으로 정상 데이터만 필터링합니다.
df_cleaned = df[valid_mask].reset_index(drop=True)

print(f"전처리 전 데이터 크기: {df.shape}")
print(f"전처리 후 데이터 크기: {df_cleaned.shape}")

# 결과를 확인하기 위해 df_cleaned를 사용합니다.
df_cleaned.head()
```

```python
import json
import xml.etree.ElementTree as ET
from urllib.parse import quote, urlencode
from urllib.request import Request, urlopen
import pandas as pd

# 1. 필수 변수 정의 (에러 방지용)
API_KEY = "5c9d733a167dc2b234ae7062a28938d853e92953c9b9998a5071373d5b7eb35e"
END_POINT = "https://apis.data.go.kr/1400377/mtweather"

url = f"{END_POINT}/mountListSearch"

# 2. 파라미터 인코딩 (dataType을 XML로 변경)
query_params = urlencode({
    "numOfRows": 500,
    "pageNo": 1,
    "dataType": "XML"
})
full_url = f"{url}?serviceKey={quote(API_KEY, safe='')}&{query_params}"

try:
    # 3. urllib를 사용하여 XML 데이터 요청 및 가져오기
    req = Request(full_url, headers={'User-Agent': 'Mozilla/5.0'})
    with urlopen(req) as response:
        xml_data = response.read().decode('utf-8')

    # 4. XML 데이터 파싱
    root = ET.fromstring(xml_data)

    # 5. 강원도 관측소 필터링
    stations = []
    for item in root.iter("item"):
        obsname = item.findtext("obsname", "")
        obsid = item.findtext("obsid", "")
        localarea = item.findtext("localarea", "")

        # 지정한 키워드가 관측소 이름(obsname)에 포함되어 있는지 확인
        keywords = ["가평", "강릉", "고성", "국립춘천숲체원", "동해", "삼척", "양구", "양양", "영월", "원주", "인제", "정선", "철원", "철원남북산림협력센터", "춘천", "태백", "평창", "홍천", "화천", "횡성"]
        if any(k in obsname for k in keywords):
            stations.append({"obsid": obsid, "obsname": obsname, "localarea": localarea})

    # 6. 데이터프레임 변환 및 출력
    df_stations = pd.DataFrame(stations)
    print(f"필터링된 관측소 개수: {len(df_stations)}개")
    print(df_stations)

except Exception as e:
    print(f"에러가 발생했습니다: {e}")
```

```python
import pandas as pd
import requests

AUTH_KEY = "VsvG3kWpQQeLxt5FqUEHTQ"
stn_url = "https://apihub.kma.go.kr/api/typ01/url/stn_inf.php"

params = {
    "inf": "AWS",
    "stn": "",
    "tm": "202401010000",
    "help": "0",
    "authKey": AUTH_KEY,
}

print("📡 기상청 서버에서 관측소 데이터를 정밀 파싱 중...")
res = requests.get(stn_url, params=params)
lines = res.text.split("\n")

stations = []
for line in lines:
    if line.startswith("#") or line.strip() == "" or "7777" in line:
        continue

    parts = line.split()
    if len(parts) >= 5:
        # 헤더 텍스트 라인 건너뛰기
        if "STN" in parts[0] or "LON" in parts[1]:
            continue

        stn_id = parts[0]

        # 💡 핵심 치트키: parts 안에서 숫자로 변환 가능한 소수점(위경도) 데이터를 자동으로 추출합니다.
        floats = []
        name = ""
        for p in parts[1:]:
            try:
                floats.append(float(p))
            except ValueError:
                # 숫자가 아닌 문자열이 나오면 그걸 지점 이름으로 간주합니다.
                if len(p) >= 2 and not p.startswith("-"):
                    name = p

        # 위경도 좌표(소수점 2개)와 이름이 정상적으로 추출되었을 때만 추가
        if len(floats) >= 2:
            # 보통 경도(120~130번대), 위도(30~40번대) 순서 혹은 반대로 들어옵니다.
            # 값의 범위를 보고 경도(lon)와 위도(lat)를 안전하게 판별합니다.
            val1, val2 = floats[0], floats[1]
            lon = val1 if val1 > val2 else val2
            lat = val2 if val1 > val2 else val1

            # 만약 위 로직에서 이름을 못 찾았으면 맨 마지막 요소를 이름으로 씁니다.
            if not name:
                name = parts[-1]

            stations.append(
                {"stn_id": stn_id, "lat": lat, "lon": lon, "name": name}
            )

print("-" * 60)

# 2. 안전장치 후 데이터프레임 생성 및 강원도 필터링
if stations:
    df_stn = pd.DataFrame(stations)

    gangwon = df_stn[
        (df_stn["lat"] >= 37.0)
        & (df_stn["lat"] <= 38.6)
        & (df_stn["lon"] >= 127.5)
        & (df_stn["lon"] <= 129.4)
    ].reset_index(drop=True)

    print(f"🎉 성공! 강원도 AWS 관측소 개수: {len(gangwon)}개")
    print("-" * 60)
    display(gangwon[["stn_id", "name", "lat", "lon"]])
else:
    print("😢 여전히 파싱된 데이터가 없습니다. 서버 응답 메시지를 확인해봐야 합니다.")
    print("기상청 응답 실제 데이터 3줄 예시:")
    valid_lines = [l for l in lines if l.strip() and not l.startswith("#")]
    print("\n".join(valid_lines[:3]))
```

```python
import os
import time
from datetime import datetime
import numpy as np
import pandas as pd
import requests

# 1. 기본 설정
AUTH_KEY = "VsvG3kWpQQeLxt5FqUEHTQ"
BASE_URL_DAY = "https://apihub.kma.go.kr/api/typ01/url/awsh.php"

target_stns = {"가리왕산": ["563", "217", "597"]}
all_target_stns = set([stn for stns in target_stns.values() for stn in stns])

start_time = time.time()

# 2. 안전한 API 호출 함수 (타임아웃 30초 대기)
def safe_get(params, max_retry=5):
    for attempt in range(max_retry):
        try:
            res = requests.get(BASE_URL_DAY, params=params, timeout=30)
            if res.status_code == 200:
                return res
            elif res.status_code == 403:
                print("  [403 에러] 인증키 권한이 없거나 차단되었습니다.")
                return None
        except Exception as e:
            print(f"  재시도 {attempt+1}/{max_retry} - 에러 원인: {e}")
            time.sleep(5 * (attempt + 1))
    return None

# 3. 각 산(타워)별 데이터 수집 시작
for tower, stns in target_stns.items():

    # 💡 [핵심 수정] '../'를 './'로 변경하여 코랩 작업공간 내에 안전하게 폴더를 만듭니다.
    save_dir = "./data/processed"
    save_path = f"{save_dir}/aws_{tower}.csv"

    # 폴더가 존재하지 않으면 코랩에 강제로 생성합니다.
    if not os.path.exists(save_dir):
        os.makedirs(save_dir, exist_ok=True)
        print(f"📁 안전한 저장 폴더를 생성했습니다: {save_dir}")

    # 이어받기 로직 (기존에 받다 끊긴 파일이 있으면 그 다음부터 시작)
    if os.path.exists(save_path):
        existing = pd.read_csv(save_path)
        done_dates = set(existing["date"].astype(str))
        tower_data = existing.to_dict("records")
        print(f"[{tower}] 이어서 시작 ({len(done_dates)}일 완료됨)")
    else:
        done_dates = set()
        tower_data = []
        print(f"\n[{tower}] 수집 시작...")

    # 날짜별 루프 (미리 만들어두신 date_list 사용)
    for i, date in enumerate(date_list):
        if str(date) in done_dates:
            continue

        # 단 1번 호출로 전국 데이터를 가져와서 속도를 9배 올립니다.
        params = {
            "tm": date,
            "stn": "",
            "help": "0",
            "authKey": AUTH_KEY,
        }

        res = safe_get(params)

        if res and res.text:
            lines = res.text.split("\n")
            stn_vals = {stn: [] for stn in stns}

            # 기상청 데이터 텍스트 한 줄씩 분석
            for line in lines:
                if line.startswith("#") or "7777" in line or line.strip() == "":
                    continue

                parts = line.split()
                if len(parts) >= 10:
                    current_stn = parts[1]

                    # 타겟 지점(예: 가리왕산 3개 지점) 데이터만 쏙 빼냅니다.
                    if current_stn in stn_vals:
                        try:
                            ta = float(parts[4])  # 기온 (TA)
                            hm = float(parts[5])  # 습도 (HM)
                            # 에러값이 아닌 정상 범위 데이터만 수집
                            if ta > -50 and hm >= 0:
                                stn_vals[current_stn].append(ta)
                        except ValueError:
                            pass

            # 수집된 타겟 지점들의 평균 기온 계산
            all_station_vals = []
            for s in stns:
                if stn_vals[s]:
                    all_station_vals.extend(stn_vals[s])

            day_row = {
                "date": date,
                "TA_mean": np.mean(all_station_vals) if all_station_vals else np.nan,
            }
            tower_data.append(day_row)

        # 기상청 서버 디도스 차단 방지용 미세 휴식
        time.sleep(0.2)

        # 10일(또는 10개 시간포인트)마다 중간 저장 파일 덮어쓰기
        if (i + 1) % 10 == 0:
            pd.DataFrame(tower_data).to_csv(save_path, index=False, encoding="utf-8-sig")
            elapsed = time.time() - start_time
            print(f"  {tower} | {i+1}/{len(date_list)} 완료 | 경과 시간: {elapsed/60:.1f}분")

    # 산 하나 분량이 모두 끝나면 최종 저장
    df_tower = pd.DataFrame(tower_data)
    df_tower.to_csv(save_path, index=False, encoding="utf-8-sig")
    print(f"[{tower}] 완료 ➡️ 파일 저장 완료 ({save_path})")

print(f"\n✨ 전체 수집 완료! 총 소요시간: {(time.time()-start_time)/60:.1f}분")
```

```python
import requests
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

session = requests.Session()
retry = Retry(total=5, backoff_factor=1)
adapter = HTTPAdapter(max_retries=retry)
session.mount("https://", adapter)

def safe_get(params, max_retry=5):
    for attempt in range(max_retry):
        try:
            res = session.get(BASE_URL_DAY, params=params, timeout=10)
            return res
        except Exception as e:
            print(f"  재시도 {attempt+1}/{max_retry} - {e}")
            time.sleep(3 * (attempt + 1))
    return None

print("safe_get 정의 완료")
```

```python
import os

all_data = {}
total_calls = sum(len(v) for v in target_stns.values()) * len(date_list) * len(obs_list)
call_count = 0
start_time = time.time()

for tower, stns in target_stns.items():
    save_path = f"../data/processed/aws_{tower}.csv"

    # 중간 저장본 있으면 이어서 시작
    if os.path.exists(save_path):
        existing = pd.read_csv(save_path)
        done_dates = set(existing["date"].astype(str))
        tower_data = existing.to_dict("records")
        print(f"[{tower}] 이어서 시작 ({len(done_dates)}일 완료됨)")
    else:
        done_dates = set()
        tower_data = []
        print(f"\n[{tower}] 수집 시작...")

    for i, date in enumerate(date_list):
        if date in done_dates:
            call_count += len(stns) * len(obs_list)
            continue

        day_row = {"date": date}

        for obs in obs_list:
            vals = []
            for stn in stns:
                params = {
                    "tm2": date,
                    "obs": obs,
                    "stn": stn,
                    "disp": "0",
                    "help": "0",
                    "authKey": AUTH_KEY
                }
                res = safe_get(params)
                if res:
                    for line in res.text.split("\n"):
                        if line.startswith("#") or "7777" in line or line.strip() == "":
                            continue
                        parts = line.split()
                        if len(parts) >= 6:
                            try:
                                vals.append(float(parts[5]))
                            except:
                                pass
                call_count += 1
                time.sleep(0.2)

            day_row[obs] = np.mean(vals) if vals else np.nan

        tower_data.append(day_row)

        # 10일마다 중간 저장
        if (i + 1) % 10 == 0:
            pd.DataFrame(tower_data).to_csv(save_path, index=False, encoding="utf-8-sig")
            elapsed = time.time() - start_time
            progress = call_count / total_calls
            eta = (elapsed / progress - elapsed) if progress > 0 else 0
            print(f"  {tower} | {i+1}/{len(date_list)}일 | "
                  f"진행률 {progress*100:.1f}% | "
                  f"경과 {elapsed/60:.1f}분 | "
                  f"남은시간 약 {eta/60:.1f}분")

    all_data[tower] = pd.DataFrame(tower_data)
    all_data[tower].to_csv(save_path, index=False, encoding="utf-8-sig")
    print(f"[{tower}] 완료 → 저장됨")

print(f"\n전체 완료 | 총 소요시간: {(time.time()-start_time)/60:.1f}분")
```
